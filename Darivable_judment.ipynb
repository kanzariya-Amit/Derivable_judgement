# Import Libraries

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

# Load Dataset

```python
df = pd.read_csv("health_dataset_500_rows.csv")
df.head()
```

# Hypothesis Formulation

## Hypothesis 1: Smoking vs Diabetes

- H₀: Smoking has no effect on diabetes
- H₁: Smoking affects diabetes

## Hypothesis 2: Age vs Blood Pressure
- H₀: Age has no effect on blood pressure
- H₁: Age affects blood pressure

# Calculate Confidence Intervals 

```python
mean_age = df['age'].mean()
std_age = df['age'].std()
n = len(df)

z = stats.norm.ppf(0.975)

margin_error = z * (std_age / np.sqrt(n))

lower = mean_age - margin_error
upper = mean_age + margin_error

print("Confidence Interval (Age):", (lower, upper))
```

# Find Critical Value and p-value

```python
z_critical = stats.norm.ppf(0.975)
print("Critical Value (Z):", z_critical)

sample_mean = df['bmi'].mean()
pop_mean = 25
std = df['bmi'].std()
n = len(df)

z_score = (sample_mean - pop_mean) / (std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
```

# Perform Z-test and T-test (Sample Mean Comparison)

```python
mean = df['bmi'].mean()
std = df['bmi'].std()
n = len(df)

pop_mean = 25

z_score = (mean - pop_mean) / (std / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
```

# Chi-Square Test (Smoking vs Diabetes)

```python
contingency_table = pd.crosstab(df['smoking_status'], df['diabetes'])

chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-square value:", chi2)
print("P-value:", p_value)
```

# ANOVA Test (Age Group vs BMI)

```python
groups = df.groupby('age_group')['bmi'].apply(list)

f_stat, p_value = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

# Covariance and Correlation (Age vs BMI)

```python
covariance = df[['age', 'bmi']].cov()
print("Covariance:\n", covariance)

correlation = df[['age', 'bmi']].corr()
print("Correlation:\n", correlation)
```

# Result Interpretation (Decision Rule)

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
